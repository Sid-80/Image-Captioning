```python
inception_v3 = tf.keras.applications.InceptionV3(
    weights='imagenet',
    include_top=False #TO remove topmost layer
)
```

```python
def load_img(img_path):
    img = tf.io.read_file(img_path)
    img = tf.io.decode_jpeg(img, channels=3)
    img = tf.keras.layers.Resizing(299,299)(img) #load as tensor
    img = img/255
    return img

def get_feat_vectors(img_path):
    img = load_img(img_path)
    img = tf.expand_dims(img,axis=0) #batch axis
    feat_vector = inception_v3(img)
    return img,feat_vector
```

```python
captions = pd.read_csv('captions.txt')
captions.head()
```

```python
def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub('\s+', ' ', text)
    text = text.strip()
    text = '[start] ' + text + ' [end]'
    return text

captions['caption'] = captions['caption'].apply(preprocess)
captions['image'] = captions['image'].apply(
    lambda x: f'/Users/siddharth/Deep Learning/Image Captioning/images/{x}')
captions.head()
```

```python
MAX_LENGTH = 40
VOCABULARY_SIZE = 10000
BATCH_SIZE = 32
BUFFER_SIZE = 1000
EMBEDDING_DIM = 512
UNITS = 512
```

```python
tokenizer = tf.keras.layers.TextVectorization(
    max_tokens=VOCABULARY_SIZE,
    standardize=None,
    output_sequence_length=MAX_LENGTH)
```

```python
tokenizer.adapt(captions['caption']) #analyzes the dataset to build the vocabulary and determines the frequency of each word.
```

```python
# This layer maps strings (words) to integer indices.
word2idx = tf.keras.layers.StringLookup(
    mask_token="",
    vocabulary=tokenizer.get_vocabulary())

#Instead of mapping strings to indices, it maps indices back to strings (words). This is useful for decoding sequences back to human-readable text.
idx2word = tf.keras.layers.StringLookup(
    mask_token="",
    vocabulary=tokenizer.get_vocabulary(),
    invert=True)
```

```python
#Creating a Mapping from Images to Captions
img_to_cap_vector = collections.defaultdict(list)
for img, cap in zip(captions['image'], captions['caption']):
    img_to_cap_vector[img].append(cap)

img_keys = list(img_to_cap_vector.keys())
random.shuffle(img_keys)

#Shuffling and Splitting the Data
slice_index = int(len(img_keys)*0.8)
img_name_train_keys, img_name_val_keys = (img_keys[:slice_index], 
                                          img_keys[slice_index:])

train_imgs = []
train_captions = []
for imgt in img_name_train_keys:
    capt_len = len(img_to_cap_vector[imgt]) #get captions associated with corresponding image
    train_imgs.extend([imgt] * capt_len) 
    train_captions.extend(img_to_cap_vector[imgt])

val_imgs = []
val_captions = []
for imgv in img_name_val_keys:
    capv_len = len(img_to_cap_vector[imgv])
    val_imgs.extend([imgv] * capv_len)
    val_captions.extend(img_to_cap_vector[imgv])
```

```python
def load_data(img_path, caption):
    img = tf.io.read_file(img_path)
    img = tf.io.decode_jpeg(img, channels=3)
    img = tf.keras.layers.Resizing(299, 299)(img)
    img = img / 255.
    caption = tokenizer(caption)
    return img, caption
```
