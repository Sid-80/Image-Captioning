```python
inception_v3 = models.inception_v3(weights='DEFAULT')
print(inception_v3)
```

```python
# We want to remove the topmost layer
# We need the vectors of images
class Identity(nn.Module):
    def __init__(self):
        super().__init__()
        
    def forward(self, x):
        return x

inception_v3.fc = Identity()
print(inception_v3)
```

```python
# Transform to preprocess the image
preprocess = transforms.Compose([
    transforms.Resize((299, 299)),  # Resize to 299x299
    transforms.ToTensor(),          # Convert to tensor
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  # Normalize
])
```

```python
def load_img(img_path):
    img = Image.open(img_path).convert('RGB')  # Load image and convert to RGB
    img = preprocess(img)  # Apply preprocessing
    return img

def get_feat_vectors(img_path):
    img = load_img(img_path)
    img = img.unsqueeze(0)  # Add batch dimension
    with torch.no_grad():
        feat_vector = inception_v3(img)
    return img, feat_vector
```

## Preprocessing the captions

```python
captions = pd.read_csv('captions.txt')
captions.head()
```

```python
def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub('\s+', ' ', text)
    text = text.strip()
    text = '[start] ' + text + ' [end]'
    return text

captions['caption'] = captions['caption'].apply(preprocess)
captions['image'] = captions['image'].apply(
    lambda x: f'/Users/siddharth/Deep Learning/Image Captioning/images/{x}')
captions.head()
```

```python
from transformers import BertTokenizer
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
import pandas as pd
from collections import defaultdict
import random

# Initialize tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Tokenization and padding function
def tokenize_and_pad(caption, max_length=MAX_LENGTH):
    tokens = tokenizer.encode(caption, add_special_tokens=True, max_length=max_length, padding='max_length', truncation=True)
    return tokens


# Create a mapping from images to captions
img_to_cap_vector = defaultdict(list)
for img, cap in zip(captions['image'], captions['caption']):
    img_to_cap_vector[img].append(cap)

img_keys = list(img_to_cap_vector.keys())
random.shuffle(img_keys)

# Split data into training and validation sets
slice_index = int(len(img_keys) * 0.8)
img_name_train_keys, img_name_val_keys = img_keys[:slice_index], img_keys[slice_index:]

# Prepare training and validation data
train_imgs = []
train_captions = []
for imgt in img_name_train_keys:
    capt_len = len(img_to_cap_vector[imgt])
    train_imgs.extend([imgt] * capt_len)
    train_captions.extend(img_to_cap_vector[imgt])

val_imgs = []
val_captions = []
for imgv in img_name_val_keys:
    capv_len = len(img_to_cap_vector[imgv])
    val_imgs.extend([imgv] * capv_len)
    val_captions.extend(img_to_cap_vector[imgv])
```

```python
preprocess = transforms.Compose([
    transforms.Resize((299, 299)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])
```

```python
class ImageCaptionDataset(Dataset):
    def __init__(self, img_paths, captions, tokenizer, transform=None):
        self.img_paths = img_paths
        self.captions = captions
        self.tokenizer = tokenizer
        self.transform = transform

    def __len__(self):
        return len(self.img_paths)

    def __getitem__(self, idx):
        img_path = self.img_paths[idx]
        caption = self.captions[idx]

        img = Image.open(img_path).convert('RGB')
        if self.transform:
            img = self.transform(img)

        caption = torch.tensor(tokenize_and_pad(caption), dtype=torch.long)
        return img, caption
```
